```python
# Mean,Median & Mode from grouped Data
"""One of your classmates has collected the numbers of human beings suffering from 
cholera and typhoid, belonging to different age groups present in certain locality. Write a python 
program for computing the mean median mode of following grouped data. 
age_groups = [(15, 25), (25, 35), (35, 45), (45, 55), (55, 65)]
frequencies = [350, 375, 120, 80, 90]"""
```

```python
age_groups = [(15, 25), (25, 35), (35, 45), (45, 55), (55, 65)]
frequencies = [350, 375, 120, 80, 90]

#Total Frequency
N=sum(frequencies)

#Midpoints calculation from age_groups
mid_points=[(l+u)/2 for l,u in age_groups]

#Mean calculation
fx=[f*x for f,x in zip(mid_points,frequencies)]
mean=sum(fx)/N

#Calculationg cummulative frequency
cumm_freq=[sum(frequencies[:i+1]) for i in range(len(frequencies))]

#Calculation Median
median_class=next(cf for cf in cumm_freq if cf>=N/2)             #output 725 median class from cumm_freq checking N/2=507.5>[350,725,845,925,1015]
F_idx=cumm_freq.index(median_class)                              #Finding the index of 725 from cumm_freq i.e 1
L=age_groups[F_idx][0]                              #finding lower boundry at 1st index from age_groups (25,35) then 0 index value i.e 25
f=frequencies[F_idx]                               #Getting frequency against the cumm_frequency (output 375) 
F=frequencies[F_idx-1]                              #Frequency preceding the f value (output 350 which is just above 375)
h=age_groups[F_idx][1]-age_groups[F_idx][0]         # h is the diff between higher boundary and lower boundry (output 35-25=10)
median = L + ((N/2 - F) / f) * h                   #Median formula to calculate median from grouped data

#Calculation of Mode
modal_class_idx=frequencies.index(max(frequencies))     #Getting the index number for the max number in frequency (output 1)
L_mode=age_groups[modal_class_idx][0]                   #finding lower boundry against the max frequency (output 1)
f1=frequencies[modal_class_idx]                                         #output 375
f0=frequencies[modal_class_idx-1]                                       #output 350
f2=frequencies[modal_class_idx+1]                                       #output 120
H_mode=age_groups[modal_class_idx][1]-age_groups[modal_class_idx][0]    
mode = L_mode + ((f1 - f0) / ((2 * f1) - f0 - f2)) * H_mode
#Calculation for SD & Variance

# Calculate squared deviations
sd=[(x-mean)**2 for x in mid_points]
f_sd=[a*b for a,b in zip(frequencies,sd)]
variance=sum(f_sd)/N
std_dev=variance**0.5
print(f"age_groups\tf\tx\tfx\t\tcf\tsd(x-mean)^2\tsd*f")
for i in range(len(frequencies)):
    print(f"{age_groups[i]}\t{frequencies[i]}\t{mid_points[i]}\t{fx[i]}\t\t{cumm_freq[i]}\t{round(sd[i],2)}\t\t{f_sd[i]}")
print(f"Total\t\t{N}\t\t{sum(fx)}\t\t\t\t\t{sum(f_sd)}")
print(f"Mean={sum(fx)}/{N}={round(mean,2)}\nMedian={round(median,2)}\nMode={round(mode,2)}\
      \nVariance={round(sum(f_sd),2)}/{N}={round(variance,2)}\nStandard Deviation={round(std_dev)}")
```
