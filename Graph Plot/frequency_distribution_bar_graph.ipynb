```python
#Frequency distribution bar chart using matplotlib

import matplotlib.pyplot as plt                                 # Import the matplotlib library for plotting
data = [2,2,3,3,3,4,4,4,5,6,6,6,7,7,7,7,8,2,4,5,6,8,8,9]        # Define the dataset
freq = {}                                                       # Create an empty dictionary to store frequencies
for i in data:                                                  # Count the frequency of each unique value in the data
    freq[i] = freq.get(i, 0) + 1  
print(freq)
classes = list(freq.keys())                                     # Extract the unique values (classes) from the dictionary
classes = sorted(freq)                                          # Sort the classes to ensure the x-axis is ordered
frequency = list(freq.values())                                 # Extract the frequency values
print(classes)
print(frequency)
plt.figure(figsize=(10, 4))                                     # Create a figure with specified size
plt.bar(classes, frequency, color="green", edgecolor="red")     # Plot a bar chart with green bars and red edges
plt.xlabel("Value")                                             # Label the x-axis
plt.ylabel("Frequency")                                         # Label the y-axis
plt.title("Frequency Distribution")                             # Add a title to the chart
for i, val in enumerate(frequency):                             # Add frequency labels above each bar
    plt.text(classes[i], val + 0.05, str(val), ha='center')     # Position label slightly above bar
plt.show()                                                      # Display the plot
```
