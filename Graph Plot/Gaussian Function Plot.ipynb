```python
import numpy as np
import matplotlib.pyplot as plt
mu=0                                                                    # Mean
sigma=1                                                                 #Standard Deviation
x=np.linspace(-4*sigma,4*sigma,1000)                                    # Range of x values
y=(1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(x-mu)**2/(2*sigma**2))          #Gaussian Formula

#Plot the function

plt.figure(figsize=(8,5))                                   
plt.plot(x,y,color='green',linewidth=2)
plt.title("Gaussian Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
