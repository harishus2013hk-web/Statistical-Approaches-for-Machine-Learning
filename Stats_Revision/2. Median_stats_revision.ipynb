```python
#Calculate the median for the given list
dataset=[1,4,8,9,11,9]
dataset.sort()                                  #Sorting the dataset before calculation
l=len(dataset)                                  #Output is 6 (Even number)
if l%2==0:                                      #Satisfies even condition
    med=(dataset[l//2]+dataset[l//2-1])/2       #med=(9+8)/2=17/2=8.5
else:
    med=dataset[l//2]
print(f"Median of \033[1m\033[31m{dataset} \033[0mis : \033[1m\033[32m{med}")

#\033[1m -> To bold the text
# \033[31m] -> To colour the text (31m, 32m, 33m ...)
# \033[0m -> To Reset the formatting
        
```
