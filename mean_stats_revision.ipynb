```python
list=[1,4,7,9]
s=sum(list)
L=len(list)
print(f"Sum of list is: {s}\nLength of list is: {L}")
```

```python
P=[24,73,56,28,19,18,17,34,71]
s=sum(P)
c=len(P)
mean_p=s/c
print(f"Sum of {P}: {s} \nTotal numbers: {c}\nMean: {round(mean_p,2)}")
```
